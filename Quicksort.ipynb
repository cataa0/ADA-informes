# 1.Descripcion del problema

**Entrada**: Secuencia de n números $[a_1,a_2,...,a_n]$

**Salida**: Permutación ordenada de la secuencia de entrada: $[a_1',a_2',...,a_n']$, de tal manera que $a_1'\leq a_2' \leq... \leq a_n'$.

El Quicksort al igual que el MergeSort es un algoritmo de divide y venceras
el cual tiene como objetivo ordenar de menor a mayor una secuencia de numeros, en este algoritmo se elige un pivote y despues se divide el arreglo en 2 sub-arreglos.

# 2.Descripcion del Algoritmo

1.Se elige un numero del arreglo el cual se le llama pivote.
2.Se ordena el arreglo de modo que a la izquierda esten los menores al oivote y a la derecha los mayores al pivote.
3.Despues de ordenarlo el pivote estaria centrado donde deberia.
4.Se repite esto de manera recursiva en cada subarreglo hasta tener el arreglo ordenado.
